```python
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
dataTrain = pd.read_csv("extendedTrain.csv")
X_test = pd.read_csv("extendedTest.csv")
```

```python
dataTrain = dataTrain.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
dTCols = dataTrain.columns.tolist()
y_train = dataTrain['TARGET']
X_train = dataTrain.drop('TARGET', axis=1)
```

```python
X_test = X_test.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

Gradient Boosting

```python
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
#from sklearn.decomposition import PCA
```

```python
gb_model = HistGradientBoostingClassifier(loss='log_loss', learning_rate = 2.5, max_iter = 3000, random_state=42)
```

```python
gb_model.fit(X_train, y_train)
```

```python
predictions = gb_model.predict_proba(X_test[X_train.columns])[:, 1]
```

```python
cv = StratifiedKFold(n_splits = 3, shuffle=True, random_state=42)
precision = cross_val_score(gb_model, X_train, y_train, cv=cv, scoring='precision')
recall = cross_val_score(gb_model, X_train, y_train, cv=cv, scoring='recall')
```

```python
print("Precision score:", precision)
```

```python
print("Recall score:", recall)
```

```python
output = pd.DataFrame({'SK_ID_CURR': X_test['SK_ID_CURR'].astype(int), 'TARGET': predictions})
output.to_csv('OUTPUT_GB.csv', index=False)
```
